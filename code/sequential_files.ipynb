```python
from xgboost import XGBClassifier
import torch
from torch.utils.data import DataLoader, Dataset

import torch.nn as nn

# Define a custom dataset for a list of dataframes
class SequentialDataset(Dataset):
    def __init__(self, dataframes):
        self.data = pd.concat(dataframes, ignore_index=True)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return torch.tensor(self.data.iloc[idx].values, dtype=torch.float32)

# Define a simple Transformer model
class TransformerModel(nn.Module):
    def __init__(self, input_dim, model_dim, num_heads, num_layers, output_dim):
        super(TransformerModel, self).__init__()
        self.embedding = nn.Linear(input_dim, model_dim)
        self.transformer = nn.Transformer(model_dim, num_heads, num_layers)
        self.fc = nn.Linear(model_dim, output_dim)

    def forward(self, x):
        x = self.embedding(x)
        x = self.transformer(x, x)
        x = self.fc(x)
        return x

# Prepare the dataset and dataloader
# dataframes = [pd.read_parquet(f'../child-mind-institute-problematic-internet-use/series_train.parquet/id={id}/part-0.parquet') for id in ids]

# # Print the dimensions of the dataframes
# for df in dataframes:
#     print(df.shape)

for id in ids:
    dataframe = pd.read_parquet(f'../child-mind-institute-problematic-internet-use/series_train.parquet/id={id}/part-0.parquet')
    print(dataframe.shape)
```
